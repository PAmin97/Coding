```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

```python
target = new_client.target
target_query = target.search('aromatase')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
selected_target = targets.target_chembl_id[0]
selected_target
```

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type='IC50')
```

```python
df = pd.DataFrame.from_dict(res)
#df.head(3)
```

```python
df.standard_type.unique()
```

```python
df.to_csv('aromatase_bioactivity_data.csv', index = False)
```

```python
df = pd.read_csv('aromatase_bioactivity_data.csv')
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df2 = df[df.standard_value.notna()]
df2.head()
```

```python
df2.isnull().sum()
```

```python
bioactivity = []
for i in df2.standard_value:
  if float(i) >= 10000:
    bioactivity.append("inactive")
  elif float(i) <= 1000:
    bioactivity.append("active")
  else:
    bioactivity.append("intermediate")
```

```python
df2['Bioactivity'] = bioactivity
df2.head()
```

```python
df3 = df2[['molecule_chembl_id', 'canonical_smiles', 'Bioactivity', 'standard_value']].copy()
df3.head()
```

```python
pd.concat([df3,pd.Series(bioactivity)], axis=1)
```
