```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://finance.yahoo.com/quote/AAPL/history/'
```

```python
r = requests.get(url)
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
history = soup.find_all('tr', attrs={'class': 'BdT Bdc($seperatorColor) Ta(end) Fz(s) Whs(nw)'})
```

```python
h = soup.find_all('tbody')
h
```

```python
Date = []
Open = []
High = []
Low = []
Close = []
Adj_Close = []
Volume = []
```

```python
for i in h:
    row = i.find_all('tr')
    print(row[1].text.strip())
```

```python
for i in h:
    row = i.find_all('tr')
    Date.append(row[0].text.strip())
    Open.append(row[1].text.strip())
    High.append(row[2].text.strip())
    Low.append(row[3].text.strip())
    Close.append(row[4].text.strip())
    Adj_Close.append(row[5].text.strip())
    Volume.append(row[6].text.strip())
```
